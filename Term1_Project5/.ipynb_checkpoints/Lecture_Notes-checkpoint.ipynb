### Draw a rectangle

```python
import numpy as np
import cv2
import matplotlib.image as mimg
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def draw_rectange(img,bbox, color=(0,0,255), thickness=3):
    draw_img = img.copy()
    
    for box in bbox:
        start = box[0]
        end = box[1]
        cv2.rectangle(draw_img,start,end,color=color, thickness=thickness)
    
    return draw_img
```

```python
img = mimg.imread("../Term1_Project4/test_images/test6.jpg")
bbox = [((800,400),(950,500)),((1000,400),(1200,500))]
draw_img = draw_rectange(img,bbox)
plt.imshow(draw_img)
```
